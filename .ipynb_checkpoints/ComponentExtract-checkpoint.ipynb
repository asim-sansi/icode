```python
import cv2
```

```python
image=cv2.imread("resources/stack-semi.png")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (12, 3))
dilation = cv2.dilate(thresh1, rect_kernel, iterations=1)
```

```python

 # Storing height and width of image
hhh, www = image.shape[:2]

        # Creating initial boundary around image
cv2.rectangle(image, (0, 0), (www, hhh), (255, 255, 255), www//150)
c=cv2.Canny(image,0,100)
(contours, h) = cv2.findContours(c.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow("canny",c)
for c in contours:
    x, y, w, h = cv2.boundingRect(c)
    #if w>15 and h>15:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 1)
cv2.imshow("d",image)
cv2.waitKey()
            
    
```
